Initialize necessary functions and load the data 

```python
import os
import pyreadr
import time
import datetime
import folium
import numpy as np
import pandas as pd
import geopandas as gpd
import scipy.sparse as sp
from scipy.linalg import solve
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import matplotlib.colors as mcolors
import statsmodels.api as sm
import igraph as ig
from functools import partial
from sklearn.metrics import mean_squared_error
from folium.plugins import MarkerCluster
from adjustText import adjust_text
import matplotlib.dates as mdates
from concurrent.futures import ProcessPoolExecutor, ThreadPoolExecutor, as_completed
from joblib import Parallel, delayed
import logging
####################################################
# Set working directory
# ~~~~~~~~STAR_LASSO
print("Empirical study Analysis of Air Quality in Bavaria, Germany")
#Import the cleaned data
data = pd.read_csv(f'Data/PM10_BFF.csv')
dates = pd.to_datetime(data['Date'])
Y = data.drop('Date', axis=1).values.T
#Y = Y[:, :100]
Time = Y.shape[1]
n = Y.shape[0]
time_index = np.arange(Time)
#############################################################################################
def create_design_matrix_X(time_index, Time, n):
    # Sine components
    sine_func_yearly = np.sin((2 * np.pi / (365 * 24)) * time_index)
    sine_func_semi_yearly = np.sin((2 * np.pi / (365 * 24 / 2)) * time_index)
    sine_func_monthly = np.sin((2 * np.pi / (365 * 24 / 12)) * time_index)
    sine_func_daily = np.sin((2 * np.pi / 24) * time_index)

    # Cosine components
    cosine_func_yearly = np.cos((2 * np.pi / (365 * 24)) * time_index)
    cosine_func_semi_yearly = np.cos((2 * np.pi / (365 * 24 / 2)) * time_index)
    cosine_func_monthly = np.cos((2 * np.pi / (365 * 24 / 12)) * time_index)
    cosine_func_daily = np.cos((2 * np.pi / 24) * time_index)

    # Combine all into a design matrix
    X_single = np.column_stack((np.ones(Time),  # Intercept
        np.linspace(0, 1, Time),  # Time trend
        sine_func_yearly,sine_func_semi_yearly,sine_func_monthly,sine_func_daily,cosine_func_yearly,cosine_func_semi_yearly,cosine_func_monthly,cosine_func_daily))
    # Expand for all stations
    X = np.tile(X_single[np.newaxis, :, :], (n, 1, 1))
    
    return X
#######################################################################################################
X = create_design_matrix_X(time_index, Time, n)
k = X.shape[2]
print(f'The number of hourly timepoints are {Time} across {n} monitoring stations for {k} covariates  \n The shape of the Y is {Y.shape} and X is {X.shape}')
#########################################################################################################################
```

# Unpenalized Maximum Likelihood Estimation

```python
#############Unpenalized Maximum Likelihood Estimation ############################################
# Timing for the parameter estimation process
ULL_starttimer = pd.Timestamp.now()

# Define the Log-Likelihood function without penalization
def LL_unpenalized(parameters, Y, X, k, n, Time, active_indices):
    # Reconstruct the full parameter vector, including active parameters
    full_params = np.zeros(k + n + n*(n-1) + 1)
    full_params[active_indices] = parameters
    
    # Extract individual parameter sets
    beta = full_params[:k]
    phi = full_params[k:k+n]
    W = np.zeros((n, n))
    W[np.triu_indices(n, 1)] = full_params[int(k+n):int(k+n+(0.5*n*(n-1)))]
    W[np.tril_indices(n, -1)] = full_params[int(k+n+(0.5*n*(n-1))):int(k+(n*n))]    
    sigma2_eps = full_params[-1] + 1e-6  # Ensure positivity
    
    # Compute determinant term for log-likelihood
    try:
        det_term = np.linalg.slogdet(np.eye(n) - W)[1]
    except np.linalg.LinAlgError:
        return 1e10  # Penalize infeasible region
    
    # Compute the residuals and sum of squares
    residuals_est = np.zeros(Time - 1)
    for t in range(1, Time):
        u_t = Y[:, t] - W @ Y[:, t] - (phi * Y[:, t-1]) - X[:, t] @ beta
        residuals_est[t-1] = np.sum(u_t**2 / sigma2_eps)
    
    sum_of_squares = np.sum(residuals_est)
    
    # Calculate log-likelihood
    constant = -0.5 * (Time - 1) * (np.log(2 * np.pi) + n * np.log(sigma2_eps))
    LogLik = constant + (Time - 1) * det_term - 0.5 * sum_of_squares

    return -LogLik  # Negative for minimization
#################################################################################################
#################################################################################################
# Spectral norm constraint function for W
def spectral_norm_constraint(params, n, k, active_indices):
    # Reconstruct W matrix from parameters
    full_params = np.zeros(k + n + n*(n-1) + 1)
    full_params[active_indices] = params  # Restore only active parameters
    beta = full_params[:k]
    phi = full_params[k:k+n]
    W = np.zeros((n, n))
    W[np.triu_indices(n, 1)] = full_params[int(k+n):int(k+n+(0.5*n*(n-1)))]
    W[np.tril_indices(n, -1)] = full_params[int(k+n+(0.5*n*(n-1))):int(k+(n*n))]
    
    # Compute the spectral radius (maximum absolute eigenvalue)
    try:
        eigvals = np.linalg.eigvals(W)
        spectral_radius = max(abs(eigvals))
    except np.linalg.LinAlgError:
        return 1e10  # Penalize infeasible region if matrix is singular or nearly singular
    
    return spectral_radius  # Ensure spectral norm < 1
constraints = [{'type': 'ineq', 'fun': lambda x: spectral_norm_constraint(x, n, k,active_indices)}]
#################################################################################################
#################################################################################################
# Load optimized parameters from file
parameters_opt = np.loadtxt(f'ESULTS/optimised_parameter_X_10.txt')
active_indices = np.where(parameters_opt != 0)[0]  # Identify active parameters
init_params = parameters_opt[active_indices]  # Initialize parameters for optimization

# Set bounds for each parameter 
lb = np.concatenate(([-np.inf]*k, [0]*n, [0]*(n*(n-1)), [1e-6]))
ub = np.concatenate(([np.inf]*k, [1]*n, [1]*(n*(n-1)), [None]))
bounds = list(zip(lb, ub))
active_bounds = [bounds[i] for i in active_indices]  # Select bounds for active parameters
#################################################################################################
print("Parameter Estimation for the unpenalized model has begun")
print("Spectral norm constraint:", spectral_norm_constraint(init_params, n, k, active_indices))  # Should be < 1
# Run the optimization to minimize the log-likelihood function
resulti = minimize(LL_unpenalized, init_params, args=(Y, X, k, n, Time, active_indices),
                  method='SLSQP', bounds=active_bounds, options={'disp': True, 'maxiter': 10000}, constraints = constraints)

# Combine optimized parameters with zero parameters (for inactive ones)
final_params = np.zeros_like(parameters_opt)
final_params[active_indices] = resulti.x
# Save the optimized parameters to file
np.savetxt(f'RESULTS/Unpenalized_parameters.txt', final_params)

# End timer and output the total time taken for estimation
ULL_endtimer = pd.Timestamp.now()
print(f"Parameter Estimation for the unpenalized model has ended in {ULL_endtimer - ULL_starttimer}")
```

# NETWORK

```python
import numpy as np
import matplotlib.pyplot as plt
import igraph as ig
import pandas as pd
import matplotlib.colors as mcolors

def plot_temporal_spatial_network(parameters_opt, n, table):
    # Estimated Weight Matrix (Spatial)
    W = np.zeros((n, n))
    W[np.triu_indices(n, 1)] = parameters_opt[int(k+n):int(k+n+0.5*n*(n-1))]
    W[np.tril_indices(n, -1)] = parameters_opt[int(k+n+0.5*n*(n-1)):int(k+n*n)]
    
    # Temporal Weights
    temporal_weights = np.round(parameters_opt[k:k+n], 2)

    # Create the figure and axes for subplots
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Plot the weights matrix (Spatial) in the first subplot
    ax1 = axes[0]
    im = ax1.imshow(W, cmap='Blues', origin='upper')  # Change colormap to your preference
    ax1.set_title("Spatial Weight Matrix")
    ax1.set_xlabel("Node Index")
    ax1.set_ylabel("Node Index")
    plt.colorbar(im, ax=ax1, orientation='vertical', fraction=0.02)

    # Create igraph Graph object (representing both temporal and spatial relationships)
    graph = ig.Graph(n)
    graph.vs["name"] = table['Index']  # Spatial location as node labels
    graph.vs["temporal_weight"] = temporal_weights  # Adding temporal weight to each node

    # Add spatial edges (from W) with edge weights
    for i in range(n):
        for j in range(i+1, n):
            if W[i, j] != 0:  # Only add edge if spatial weight is non-zero
                graph.add_edge(i, j, weight=W[i, j])

    # Set up visual style for the network
    visual_style = dict()
    visual_style["vertex_size"] = 20
    visual_style["edge_width"] = 0.5

    edge_weights = [graph.es[i]["weight"] for i in range(len(graph.es))]
    #edge_norm = mcolors.Normalize(vmin=W.min(), vmax=W.max())
    edge_colors = [im.get_cmap()(im.norm(weight)) for weight in edge_weights]


    visual_style["edge_color"] = edge_colors

    # **Set node colors based on temporal weights**
    max_temporal_weight = max(temporal_weights)  # Normalize temporal weights for color scale
    norm = mcolors.Normalize(vmin=0, vmax=max_temporal_weight)
    node_colors = [plt.cm.viridis(norm(weight)) for weight in temporal_weights]  # Viridis colormap for node colors
    visual_style["vertex_color"] = node_colors

    # **Make node labels white**
    visual_style["vertex_label_color"] = "black"  # Set label color to white

    # Perform layout
    layout = graph.layout_fruchterman_reingold(weights='weight')  # Layout using edge weights

    # Plot the network (with temporal and spatial effects)
    ax2 = axes[1]
    ig.plot(graph, layout=layout, vertex_label=graph.vs["name"], target=ax2, **visual_style)
    ax2.set_title("Spatiotemporal Network")

    # Add colorbar for node colors (temporal weights)
    sm = plt.cm.ScalarMappable(cmap='viridis', norm=norm)
    sm.set_array([])  # Empty array for colorbar
    fig.colorbar(sm, ax=ax2, orientation='vertical', fraction=0.02)


    # Adjust layout to prevent overlap
    plt.tight_layout()
    #plt.savefig(f"plots/Spatiotemporal_Network.pdf", bbox_inches='tight') # Save the figure as pdf
    # Display the plot
    plt.show()

plot_temporal_spatial_network(parameters_opt, n, table)
```

# Predicted values and residuals

```python
import math
def compute_predicted_values(parameters, Y, X, k, n, Time):
    beta = parameters[:k]
    phi = parameters[k:k+n]
    W = np.zeros((n, n))
    W[np.triu_indices(n, 1)] = parameters[int(k+n):int(k+n+0.5*n*(n-1))]
    W[np.tril_indices(n, -1)] = parameters[int(k+n+0.5*n*(n-1)):int(k+n*n)]
    
    Y_pred = np.zeros_like(Y)
    
    for t in range(1, Time):
        Y_pred[:, t] = W @ Y[:, t] + (phi * Y[:, t-1]) + X[:, t] @ beta
    MSE = mean_squared_error(Y, Y_pred)
    AIC = (n*Time )* math.log(MSE) + 2 * (n*n+k+1)
    BIC = (n*Time )* math.log(MSE) + ((n*n+k+1)) * math.log(n*Time)
    return Y_pred, MSE, AIC, BIC
Y_pred, MSE, AIC, BIC = compute_predicted_values(parameters_opt, Y, X, k, n, Time)
print(f"The MSE of the model is: {MSE} and AIC: {AIC} and BIC: {BIC}")
#########################################################
####################Plot of and Model  residuals
residuals = Y - Y_pred
# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot on the first subplot
axes[0].plot(dates, np.median(Y, axis=0), label='Data', color='grey')
axes[0].plot(dates, np.median(Y_pred, axis=0), label='Predicted', color='red')
axes[0].set_title('Data vs Predicted Values')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Mean Value')
axes[0].set_yscale('log')
axes[0].legend()
axes[0].grid(True)

# Plot on the second subplot
axes[1].plot(dates, np.median(residuals, axis=0), label='Residual plot', color='grey')
axes[1].set_title('Residual plot')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Residuals')
#axes[1].set_yscale('log')
axes[1].legend()
axes[1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()
# Save as PDF
#plt.savefig('plots/Predicted_and_Resiual plots.pdf', format='pdf')

# Display the plot
plt.show()
```
